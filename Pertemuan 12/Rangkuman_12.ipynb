## Shortest path
- Path adalah urutan node antara node awal dan node akhir, di mana tidak ada node yang muncul dua kali pada jalur tersebut. Jalur mewakili rute antara titik awal dan akhir yang dipilih. ini akan menjadi himpunan simpul, p, menghubungkan simpul awal dengan simpul akhir. tidak ada simpul yang diulang di hal.
- Length of path dihitung dengan menghitung sisi-sisi penyusunnya. Dari semua opsi, jalur dengan panjang terkecil disebut jalur terpendek. Penghitungan jalur terpendek digunakan secara luas dalam algoritme teori graf, tetapi tidak selalu mudah untuk dihitung. Ada berbagai algoritme yang dapat paling populer yang dapat digunakan untuk menemukan jarak terpendek antara simpul awal dan simpul akhir adalah algoritme Djikstra.

Jika kedua alter diketahui sebagai kasus penipuan, kita dapat dengan aman mendeklarasikan simpul m sebagai penipuan juga.
```python
## Pengukuran Centrality: Eigenvector
import networkx as nx
import matplotlib.pyplot as plt

vertices = range(1, 10)
edges = [(7, 2), (2, 3), (7, 4), (4, 5), (7, 3), (7, 5), (1, 6), (1, 7), (2, 8), (2, 9)]
G = nx.Graph()
G.add_nodes_from(vertices)
G.add_edges_from(edges)
nx.draw(G, with_labels=True, node_color='y', node_size=800)
```

```python
nx.degree_centrality(G)
```

```python
nx.betweenness_centrality(G)
```

```python
nx.closeness_centrality(G)
```

```python
centrality = nx.eigenvector_centrality(G)
```

```python
sorted((v, '{:0.2f}'.format(c)) for v, c in centrality.items())
```

```python
G_fb = nx.read_edgelist("facebook_combined.txt", create_using=nx.Graph(), nodetype=int)
```

```python
print(nx.info(G_fb))
```

```python
pos = nx.spring_layout(G_fb)
betCent = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
node_color = [20000.0 * G_fb.degree(v) for v in G_fb]
node_size = [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size)
plt.axis('off')
```
