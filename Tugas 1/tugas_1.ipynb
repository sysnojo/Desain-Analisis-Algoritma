#### **1. Cari nilai KPK dari 3 dan 4**
<ul>
    <li><em>Menentukan Algoritma KPK dari 3 dan 4</em></li>
    <p>Berikut adalah algoritma program untuk menentukan KPK dari 3 dan 4 :</p>
    <ol>
        <li>Mendeklarasi/buat variabel <code>a</code> dengan nilai 3</li>
        <li>Mendeklarasi/buat variabel <code>b</code> dengan nilai 4</li>
        <li>Mendeklarasi/buat variabel <code>bil_a</code> dengan nilai variabel <code>a</code></li>
        <li>Mendeklarasi/buat variabel <code>bil_b</code> dengan nilai variabel <code>b</code></li>
        <li>Membuat perulangan selama variabel <code>a</code> di modulus dengan variabel <code>bil_b</code> tidak sama dengan 0 lakukan:<br>• pertambahan oleh variabel <code>a</code> ditambah dengan variable <code>bil_a</code></li>
        <li>Membuat perulangan selama variabel <code>b</code> di modulus dengan variabel <code>bil_a</code> tidak sama dengan 0 lakukan:<br>• pertambahan oleh variabel <code>b</code> ditambah dengan variable <code>bil_b</code></li>
        <li>Membuat kondisi jika variabel <code>a</code> sama dengan variabel <code>b</code> maka lakukan:<br>• print hasil KPK dari variabel <code>a</code> dan <code>b</code></li>
    </ol>
    <br>
    <li><em>Membuat Pseudocode program KPK dari 3 dan 4</em></li>
    <p>Berikut adalah pseudocode program untuk menentukan KPK dari 3 dan 4 :</p>
    <div style="background-color:rgba(0, 0, 0, 0.0470588); padding:10px 0;font-family:monospace;">
        A ← 3<br>
        B ← 4<br>
        BIL_A ← A<br>
        BIL_B ← B<br><br>
        <em><strong>While</strong></em> A <strong>mod</strong> BIL_B != 0:<br>
        &emsp;A ← A + BIL_A<br>
        <em><strong>While</strong></em> B <strong>mod</strong>  BIL_A != 0:<br>
        &emsp;B ← B + BIL_B<br><br>
        <em><strong>If</strong></em> A == B then<br>
        &emsp;<strong>PRINT</strong> <em>"Hasil KPK dari <strong>BIL_A</strong> dan <strong>BIL_B</strong> adalah <strong>A</strong>"</em><br> 
        <br># referensi aturan https://student.cs.uwaterloo.ca/~cs231/resources/pseudocode.pdf
</div>
    <br>
    <li><em>Hasil Uji Kode Bahasa Python</em></li>
    <p>Berikut adalah hasil uji dan kode yang diimplementasikan dalam bahasa pemrograman Python :
</ul>

```python
"""
# ilustrasi kelipatan
3 = 3, 6, 9, (12), 15, 18
4 = 4, 8, (12), 16, 20
Note: dengan tujuan ingin mencari angka yang sama di kedua kelipatan
"""

a = 3 # deklarasi angka pertama
b = 4 # deklarasi angka kedua
bil_a = a # deklarasi angka statis untuk angka pertama
bil_b = b # deklarasi angka statis untuk angka kedua

# melakukan perulangan disaat a mod bil_b != 0
while a % bil_b != 0:
    a += bil_a

# melakukan perulangan disaat b mod bil_a != 0
while b % bil_a != 0:
    b += bil_b

# jika kedua kelipatan mempunyai angka yang sama, maka ditemukannya KPK
if a == b:
    print(f"hasil KPK dari {bil_a} dan {bil_b} adalah {a}") 
```

    <li><em>Membuat Pseudocode program untuk menukar posisi dua variabel x dan y</em></li>
    <p>Berikut adalah pseudocode program untuk menukar posisi dua variabel x dan y :</p>
    <div style="background-color:rgba(0, 0, 0, 0.0470588); padding:10px 0;font-family:monospace;">
        <strong>Function</strong> TUKAR_POSISI_VAR(X, Y):<br>
        &emsp;&emsp;PIRING_1 ← X<br>
        &emsp;&emsp;PIRING_2 ← Y<br>
        &emsp;&emsp;PIRING_3 ← ""<br><br>
        &emsp;&emsp;SEBELUM ← <strong>PRINT</strong> <em>"(Sebelum ditukar)\nPiring 1 = <strong>PIRING_1</strong>\nPiring 2 = <strong>PIRING_2</strong>\nPiring 3 = <strong>PIRING_3</strong>"</em><br><br>
        &emsp;&emsp;PIRING_3 ← PIRING_1<br>
        &emsp;&emsp;PIRING_1 ← PIRING_2<br>
        &emsp;&emsp;PIRING_2 ← PIRING_3<br>
        &emsp;&emsp;PIRING_3 ← ""<br><br>
        &emsp;&emsp;SESUDAH ← <strong>PRINT</strong> <em>"(Sesudah ditukar)\nPiring 1 = <strong>PIRING_1</strong>\nPiring 2 = <strong>PIRING_2</strong>\nPiring 3 = <strong>PIRING_3</strong>"</em><br><br>
        &emsp;&emsp;<strong>RETURN PRINT</strong> <em>"<strong>SEBELUM</strong>\n<strong>SESUDAH</strong>"</em><br><br>
        <strong>TUKAR_POSISI_VARIABEL(</strong><em>"manggis", "pisang"</em><strong>)</strong>
        <br><br># referensi aturan https://student.cs.uwaterloo.ca/~cs231/resources/pseudocode.pdf

    <li><em>Hasil Uji Kode Bahasa Python</em></li>
    <p>Berikut adalah hasil uji dan kode yang diimplementasikan dalam bahasa pemrograman Python :
</ul>

```python
# fungsi tukar posisi variabel x dan y
def tukar_posisi_var(x, y):
    piring_1 = x # manggis
    piring_2 = y # pisang
    piring_3 = "" # kosong

    # kondisi sebelum penukaran
    sebelum = print(f"(Sebelum ditukar)\nPiring 1 = {piring_1}\nPiring 2 = {piring_2}\nPiring 3 = {piring_3}")

    # menukar posisi x
    piring_3 = piring_1 # menampung x pada piring 3
    # menukar posisi y
    piring_1 = piring_2 # menampung y pada piring 1
    # menukar posisi x kembali
    piring_2 = piring_3 # memindahkan variabel x ke piring_2
    # mengosongkan piring 3
    piring_3 = ""

    # kondisi sesudah penukaran
    sesudah = print(f"(Sesudah ditukar)\nPiring 1 = {piring_1}\nPiring 2 = {piring_2}\nPiring 3 = {piring_3}")

    return print(f"{sebelum}\n{sesudah}")

# memanggil fungsi tukar_posisi_var(x, y)
tukar_posisi_var("manggis", "pisang")
```

```python
# fungsi menghitung volume kerucut
def hitung_volume_kerucut(n_diameter, n_tinggi):
    pi = 3.14
    n_jari = n_diameter / 2
    return (pi * (n_jari ** 2) * n_tinggi) / 3
diameter = 5 # deklarasi diameter kerucut
tinggi = 4 # deklarasi tinggi kerucut
# memanggil fungsi hitung_volume_kerucut(n_diamater, n_tinggi)
print(f"Volume kerucut = {hitung_volume_kerucut(diameter, tinggi)}")
```
